```python
import pandas as pd
import numpy as np

np.random.seed(42)

num_shifts = 300  # hospital-scale sample
shift_dates = pd.date_range(start="2025-01-01", periods=90)

shifts = pd.DataFrame({
    "shift_id": range(1, num_shifts + 1),
    "employee_id": np.random.choice(range(1, 22), num_shifts),
    "department_id": np.random.choice(range(1, 9), num_shifts),
    "shift_date": np.random.choice(shift_dates, num_shifts),
    "scheduled_hours": 12,
    "actual_hours": np.random.choice([8, 10, 12, 14], num_shifts, p=[0.15, 0.25, 0.45, 0.15]),
    "shift_type": np.random.choice(["Day", "Night"], num_shifts, p=[0.6, 0.4])
})

# Calculate overtime
shifts["overtime_hours"] = shifts["actual_hours"].apply(lambda x: max(0, x - 12))

# Assign shift status
shifts["shift_status"] = np.where(shifts["actual_hours"] == 0, "Call-Out", "Completed")

shifts = shifts.sort_values("shift_date")

shifts.to_csv("shifts.csv", index=False)
shifts.head()
```

```python
import pandas as pd
import numpy as np

np.random.seed(123)

num_records = 300
pay_dates = pd.date_range(start="2024-01-07", periods=12, freq="W")

payroll = pd.DataFrame({
    "payroll_id": range(1, num_records + 1),
    "employee_id": np.random.choice(range(1, 22), num_records),
    "department_id": np.random.choice(range(1, 9), num_records),
    "pay_date": np.random.choice(pay_dates, num_records),
    "regular_hours": np.random.choice([32, 36, 40], num_records, p=[0.2, 0.3, 0.5]),
    "overtime_hours": np.random.choice([0, 4, 8], num_records, p=[0.6, 0.25, 0.15]),
    "hourly_rate": np.random.choice(
        [20, 22, 26, 29, 32, 36, 38, 42, 45, 50, 60],
        num_records
    )
})

# Calculate pay
payroll["regular_pay"] = payroll["regular_hours"] * payroll["hourly_rate"]
payroll["overtime_pay"] = payroll["overtime_hours"] * payroll["hourly_rate"] * 1.5
payroll["total_pay"] = payroll["regular_pay"] + payroll["overtime_pay"]

payroll = payroll.sort_values("pay_date")

payroll.to_csv("payroll.csv", index=False)
payroll.head()
```

```python
conn = sqlite3.connect("hospital_analytics.db")

query = """
SELECT
    d.department_name,
    COUNT(e.employee_id) AS employee_count
FROM employees e
JOIN departments d
    ON e.department_id = d.department_id
GROUP BY d.department_name
ORDER BY employee_count DESC;
"""

pd.read_sql_query(query, conn)
```

### Staffing Distribution Summary

This analysis shows how employees are distributed across departments within the facility.

Key Observations:
- Staffing levels vary significantly by department, reflecting differences in patient volume, acuity, and operational requirements.
- Higher headcount departments likely represent front-line or high-throughput areas (e.g., Emergency Department, Med Surge).
- Lower headcount departments may indicate specialized or support functions that require fewer staff but higher skill levels.

This metric establishes a baseline for evaluating labor costs, staffing efficiency, and reliability in subsequent analyses.

```python
query = """
SELECT
    d.department_name,
    ROUND(AVG(e.hourly_rate), 2) AS avg_hourly_rate
FROM employees e
JOIN departments d
    ON e.department_id = d.department_id
GROUP BY d.department_name
ORDER BY avg_hourly_rate DESC;
"""

pd.read_sql_query(query, conn)
```

### Why this matters

This is important in understanding the headcount and labor cost correlations. Average hourly rate helps isolate whether cost differences are driving by staffing volume or workforce mix. This also shows that higher clinical acuity or specialized skills reflect higher average hourly rates.

This metric will be used alongside staffing levels and attendance data to evaluate labor efficiency and cost optimization opportunities.

```python
query = """
SELECT
    d.department_name,
    COUNT(a.shift_date) AS scheduled_shifts,
    SUM(CASE WHEN a.attendance_type = 'Call Out' THEN 1 ELSE 0 END) AS call_outs,
    ROUND(
        SUM(CASE WHEN a.attendance_type = 'Call Out' THEN 1 ELSE 0 END) * 1.0
        / COUNT(a.shift_date),
        3
    ) AS call_out_rate
FROM attendance a
JOIN employees e
    ON a.employee_id = e.employee_id
JOIN departments d
    ON e.department_id = d.department_id
GROUP BY d.department_name
ORDER BY call_out_rate DESC;
"""

pd.read_sql_query(query, conn)
```

### Staffing Reliability by Department

This KPI measures employee call-out rates across departments, highlighting areas where staffing reliability may impact operations.

Key observations:
- Departments with higher call-out rates face increased risk of understaffing, overtime usage, and workflow disruption.
- Reliability challenges can compound labor costs and negatively affect patient experience.
- These results provide a foundation for targeted staffing strategies and proactive workforce planning.

This analysis links workforce behavior directly to operational risk and cost control.


```python
query = """
SELECT
    d.department_name,
    ROUND(SUM(a.worked_hours), 1) AS total_worked_hours,
    ROUND(SUM(a.scheduled_hours), 1) AS total_scheduled_hours,
    ROUND(
        SUM(a.worked_hours) * 1.0 / SUM(a.scheduled_hours),
        3
    ) AS utilization_rate
FROM attendance a
JOIN employees e
    ON a.employee_id = e.employee_id
JOIN departments d
    ON e.department_id = d.department_id
GROUP BY d.department_name
ORDER BY utilization_rate DESC;
"""

pd.read_sql_query(query, conn)
```

## Does staffing reliability impact patient satisfaction?

### Business logic:
If call-outs, late arrivals, and early departures increase, patient satisfaction should drop.

```python
query = """
SELECT
    d.department_name,
    COUNT(a.attendance_id) AS total_attendance_events,
    SUM(CASE WHEN a.attendance_type = 'Call-Out' THEN 1 ELSE 0 END) AS call_outs,
    ROUND(
        SUM(CASE WHEN a.attendance_type = 'Call-Out' THEN 1 ELSE 0 END) * 1.0
        / COUNT(a.attendance_id),
        3
    ) AS call_out_rate,
    ROUND(AVG(ps.satisfaction_score), 2) AS avg_patient_satisfaction
FROM attendance a
JOIN employees e
    ON a.employee_id = e.employee_id
JOIN departments d
    ON e.department_id = d.department_id
JOIN patient_surveys ps
    ON d.department_id = ps.department_id
GROUP BY d.department_name
ORDER BY call_out_rate DESC;
"""

pd.read_sql_query(query, conn)
```

### Staffing Reliability and Patient Satisfaction

This analysis examines the relationship between staffing reliability (measured by call-out rate) and average patient satisfaction across departments.

Key observations:
- The ICU and Registration departments exhibit the highest call-out rates (77.8% and 64.3% respectively), indicating frequent staffing disruptions.
- Departments with elevated call-out rates generally do not demonstrate the highest patient satisfaction scores, suggesting potential operational strain.
- The Emergency Department shows a notably lower call-out rate (15.0%) while maintaining a comparable patient satisfaction score, indicating more stable staffing coverage.
- Pharmacy and Med Surg display moderate to lower call-out rates but still show opportunities for improving patient satisfaction.

Overall, the results suggest a relationship between staffing reliability and patient experience. Departments experiencing frequent call-outs may benefit from targeted workforce interventions to stabilize coverage and protect service quality.


## Where are labor costs concentrated relative to staffing levels?

### Business logic:
High headcount + high pay = cost pressure.

```python
query = """
SELECT
        d.department_name,
    COUNT(e.employee_id) AS employee_count,
    ROUND(AVG(e.hourly_rate), 2) AS avg_hourly_rate,
    ROUND(COUNT(e.employee_id) * AVG(e.hourly_rate), 2) AS staffing_cost_index
FROM employees e
JOIN departments d
    ON e.department_id = d.department_id
GROUP BY d.department_name
ORDER BY staffing_cost_index DESC;
"""
pd.read_sql_query(query, conn)
```

### Labor Cost Concentration Across Departments

This analysis evaluates how staffing levels and average hourly pay combine to create labor cost pressure across departments.

Key observations:
- The Emergency Department represents the highest staffing cost concentration, driven by both higher headcount and elevated hourly rates.
- Med Surg also demonstrates significant cost exposure, suggesting staffing scale rather than pay rate alone contributes to labor costs.
- ICU and NICU show high average hourly rates but lower overall cost concentration due to smaller staff sizes.
- Registration maintains relatively high staffing levels but lower average pay, resulting in moderate cost pressure.
- Laboratory and Radiology exhibit the lowest staffing cost indices, indicating comparatively lower labor cost exposure.

These findings help identify where labor costs are most concentrated and where staffing adjustments may have the greatest financial impact.


## Which departments show optimization risk?

### Business logic:
Departments that are:
- High cost
- High reliability issues
- Lower patient satisfaction

are management risk zones.

```python
query = """SELECT
    d.department_name,
    COUNT(DISTINCT e.employee_id) AS employee_count,
    ROUND(AVG(e.hourly_rate), 2) AS avg_hourly_rate,
    SUM(CASE WHEN a.attendance_type = 'Call-Out' THEN 1 ELSE 0 END) AS call_outs,
    ROUND(AVG(ps.satisfaction_score), 2) AS avg_patient_satisfaction
FROM departments d
LEFT JOIN employees e
    ON d.department_id = e.department_id
LEFT JOIN attendance a
    ON e.employee_id = a.employee_id
LEFT JOIN patient_surveys ps
    ON d.department_id = ps.department_id
GROUP BY d.department_name
ORDER BY call_outs DESC, avg_patient_satisfaction ASC;
"""
pd.read_sql_query(query, conn)
```

### Operational Risk Summary by Department

This analysis consolidates staffing levels, labor costs, workforce reliability, and patient satisfaction to identify departments with elevated operational risk.

Key observations:
- Registration and ICU experience the highest volume of call-outs, indicating significant staffing reliability challenges.
- Radiology and Pharmacy display lower patient satisfaction scores relative to other departments, suggesting potential service delivery pressures.
- ICU and NICU maintain higher average hourly rates, reflecting the specialized nature of care, but differ in workforce reliability and satisfaction outcomes.
- Emergency Department and Med Surg show moderate call-out volumes with average patient satisfaction, indicating potential opportunities for operational fine-tuning rather than major restructuring.

By evaluating these factors together, leadership can better identify departments where staffing interventions may improve performance without compromising care quality.
